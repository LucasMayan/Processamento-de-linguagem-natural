```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Lucas Duarte,',
    'From': 'lucasciciri00@hotmail.com'
}

url = 'https://www.billboard.com/charts/hot-100'

page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Verifique a estrutura correta do HTML da p√°gina
lista_musicas = soup.find(class_='chart-results-list')
```

```python
lista_musicas_itens = lista_musicas.find_all(class_='c-title')
```

```python
for nome_musica in lista_musicas_itens:
    nome = nome_musica.contents[0]
    print(nome)
    if nome == "UP":
        print("Encontrei a musica na lista!!!")
```
